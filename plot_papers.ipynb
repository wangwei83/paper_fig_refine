```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter


fig, ax = plt.subplots(figsize=(10,8))

ax.set_xlabel("年份",fontsize=15)
ax.set_ylabel("文献数量（篇）",fontsize=15)

ax.plot(df_cjh.year, df_cjh.num,'b-o',label='总文献个数')
ax.plot(df_cjh_ai.year, df_cjh_ai.num,'r-o',label='AI相关文献个数')
ax.plot(df_cjh_chinese.year, df_cjh_chinese.num,'g-o',label='中文文献个数')

ax.legend(loc='upper center',fontsize=15)
# 设置 x 轴的刻度为整数
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
```
