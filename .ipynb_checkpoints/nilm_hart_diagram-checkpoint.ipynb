```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签\n
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
# http://www.getdata-graph-digitizer.com/, 数据来自HART，采用getdata-graph-digitizer
df = pd.read_csv('./NILM_hart_diagram.csv')
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(df.X,df.Y,'b-',label='功耗数据')
ax.legend(loc='upper left',fontsize=15)
ax.set_xlabel("时间（分）",fontsize=15)
ax.set_ylabel("功耗（千瓦）",fontsize=15)
plt.text(7, 0.51, r'冰箱',fontsize=12)
plt.text(16, 2.5, r'烤箱',fontsize=12)
plt.text(28, 1, r'烤箱',fontsize=12)
plt.text(28, 1.8, r'冰箱',fontsize=12)
plt.text(28, 3, r'微波炉',fontsize=12)
# 冰箱的两条线
ax.annotate("",
            xy=(2.8,0.8),
            xytext=(7,0.5),
            arrowprops=dict(arrowstyle="->", color="black"))
ax.annotate("",
            xy=(11.8,0.8),
            xytext=(8.5,0.5),
            arrowprops=dict(arrowstyle="->", color="black"))
# 烤箱的两条线
ax.annotate("",
            xy=(13,3),
            xytext=(16,2.5),
            arrowprops=dict(arrowstyle="->", color="black"))
ax.annotate("",
            xy=(19,3),
            xytext=(17.5,2.5),
            arrowprops=dict(arrowstyle="->", color="black"))

# 微波炉的两条线
ax.annotate("",
            xy=(24,2.5),
            xytext=(27.5,3),
            arrowprops=dict(arrowstyle="->", color="black"))
ax.annotate("",
            xy=(32.5,2.5),
            xytext=(30,3),
            arrowprops=dict(arrowstyle="->", color="black"))

# 冰箱的两条线
ax.annotate("",
            xy=(23.5,2),
            xytext=(27.5,1.8),
            arrowprops=dict(arrowstyle="->", color="black"))
ax.annotate("",
            xy=(35.1,0.6),
            xytext=(30,1.8),
            arrowprops=dict(arrowstyle="->", color="black"))


# 烤箱的两条线
ax.annotate("",
            xy=(22,1.1),
            xytext=(27.5,1),
            arrowprops=dict(arrowstyle="->", color="black"))
ax.annotate("",
            xy=(32.1,1),
            xytext=(30,1),
            arrowprops=dict(arrowstyle="->", color="black"))

# 冰箱开
ax.annotate("",
            xy=(2.6,0.91),
            xytext=(2.6,0.5),
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 冰箱关
ax.annotate("",
            xy=(12.1,0.66),
            xytext=(12.1,1.05),
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 烤箱开
ax.annotate("",
            xy=(12.6,3.46),
            xytext=(12.6,0.66),
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 烤箱关
ax.annotate("",
            xy=(19.38,1.38),
            xytext=(19.38,3.99),
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 烤箱开
ax.annotate("",
            xy=(22.01,1.74),
            xytext=(22.01,0.688),  
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 烤箱关
ax.annotate("",
            xy=(32.456,0.907),
            xytext=(32.456,2.085),
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 冰箱开
ax.annotate("",
            xy=(23.07,2.17),
            xytext=(23.07,1.68),
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 冰箱关
ax.annotate("",
            xy=(35.064,0.462),
            xytext=(35.064,1.497),
            arrowprops=dict(arrowstyle="-|>", color="r"))


# 微波炉开
ax.annotate("",
            xy=(24.47,2.63),
            xytext=(24.47,2.082),
            arrowprops=dict(arrowstyle="-|>", color="r"))

# 微波炉关
ax.annotate("",
            xy=(32.28070,2.067),
            xytext=(32.28070,2.639),
            arrowprops=dict(arrowstyle="-|>", color="r"))


plt.savefig("nilm_hart_diagram0923.png", dpi=600, bbox_inches='tight')
```
